```python
accidents_trend_df['yoy_change_fatal']=accidents_trend_df['no_of_fatal_accidents'].pct_change()*100
accidents_trend_df['yoy_change_fatal']=accidents_trend_df['yoy_change_fatal'].round(1)
```

```python
accidents_trend_df
```

```python
accidents_trend_df['3yr_rolling_fatal']=accidents_trend_df['no_of_fatal_accidents'].rolling(3).mean().round(1)
```

```python
accidents_trend_df['death_per_fatal']=(accidents_trend_df['no_of_persons_killed']/accidents_trend_df['no_of_fatal_accidents']).round(2)
```

```python
accidents_trend_df
```

```python
accidents_trend_df.drop('yoy_change',axis=1)
```

```python
accidents_trend_df['total_accidents']=(accidents_trend_df['no_of_fatal_accidents']+accidents_trend_df['no_of_serious_accidents'])
```

```python
accidents_trend_df['total_person_affected']=(accidents_trend_df['no_of_persons_seriously_injured']+accidents_trend_df['no_of_persons_seriously_injured_in_fatal_acc']+accidents_trend_df['no_of_persons_killed'])
```

```python
accidents_trend_df['yoy_total_accidents']=accidents_trend_df['total_accidents'].pct_change()*100
accidents_trend_df['yoy_total_accidents']=accidents_trend_df['yoy_total_accidents'].round(1)
```

```python
accidents_trend_df['person_affect_per_accident']=(accidents_trend_df['total_person_affected']/accidents_trend_df['total_accidents']).round(2)
```

```python
accidents_trend_df.drop('yoy_change',axis=1)
accidents_trend_df.to_csv('Table1_modified.csv',index=False)
```

```python
plt.figure(figsize=(15,6))
plt.plot(accidents_trend_df['year'],accidents_trend_df['no_of_fatal_accidents'],marker='o',label='Fatal Accidents')
for x, y in zip(
    accidents_trend_df['year'],
    accidents_trend_df['no_of_fatal_accidents']
):
    plt.text(x, y+2, str(y), fontsize=6)
plt.plot(accidents_trend_df['year'],accidents_trend_df['no_of_persons_killed'],marker='s',label='Person dies in Fatal Accidents')
for x, y in zip(
    accidents_trend_df['year'],
    accidents_trend_df['no_of_persons_killed']
):
 plt.text(x, y+2, str(y), fontsize=6)
plt.xticks(accidents_trend_df['year'],rotation=45)
plt.xlabel('Year')
plt.ylabel('Accidents')
plt.title('Fatal Accidents in Coal Mines from 1997 to 2022')
plt.legend()
plt.grid(True)
plt.savefig("/content/sample_data.png", dpi=150)
```

```python
plt.figure(figsize=(15,7))
plt.plot(accidents_trend_df['year'],accidents_trend_df['total_accidents'],marker='o',label='Total Accidents')
for x, y in zip(
    accidents_trend_df['year'],
    accidents_trend_df['total_accidents']
):
    plt.text(x, y+2, str(y), fontsize=6)
plt.plot(accidents_trend_df['year'],accidents_trend_df['total_person_affected'],marker='s',label='Person Affected')
for x, y in zip(
    accidents_trend_df['year'],
    accidents_trend_df['total_person_affected']
):
 plt.text(x, y+2, str(y), fontsize=6)
plt.xticks(accidents_trend_df['year'],rotation=45)
plt.xlabel('Year')
plt.ylabel('Accidents')
plt.title('Total Accidents And Total Person Affected in Coal Mines from 1997 to 2022')
plt.legend()
plt.grid(True)
plt.savefig("/content/sample_data.png", dpi=150)
```

```python
latest = accidents_placewise_long[accidents_placewise_long.year == 2022]
latest_pct = latest.copy()
latest_pct["share_pct"] = (latest_pct["fatalities"] /
                            latest_pct["fatalities"].sum() * 100).round(1)
```
